# 6 Complexity of Algorithms

## 6.1 Time Complexity 

문제의 크기와 이를 해결하는데 걸리는 시간 사이의 관계

### 6.1.1 Average Time Complexity 

임의의 입력 패턴을 가정했을 때 소요되는 시간의 평균

### 6.1.2 Worst-case Time Complexity 

가장 긴 시간을 소요하게 만드는 입력에 따라 소요되는 시간

## 6.2 Space Complexity 

문제의 크기와 이를 해결하는데 필요한 메모리 공간 사이의 관계 

단순히 주어진 배열이 아닌 문제를 해결할 때 사용하는 메모리를 모두 포괄한 개념. 

(여기선 다루지 않는다.)

## 6.3 Big-O Notation

점근 표기법 (asymptotic notation)의 하나. 

어떤 함수의 증가 양상을 다른 함수와의 비교로 표현. 

**O(x)**: x에 비례하는 시간 소요. 상수 계수를 무시한다.  

### 6.3.1 선형 시간 알고리즘 O(n)

e.g) 무작위의 배열에서 `max`, `min` 같은 것을 찾을 때 하나하나 다 살펴볼 수밖에 없음. 

### 6.3.2 로그 시간 알고리즘 O(log n)

e.g) 크기 순으로 정렬된 수에서 특정 값을 찾기 위해 이진 탐색 알고리즘을 적용할 때. 

### 6.3.3 이차 시간 알고리즘 O(n**2)

e.g) 삽입 정렬 insertion sort

- best case: O(n)
- worst case: O(n**2)

이보다 나은 복잡도를 가지는 정렬 알고리즘은? 

merge sort, O(n * log n) 

입력 패턴에 따라 속도의 차이가 있을 수 있지만 정렬 문제에 대해 O(n * log n)보다 낮은 복잡도를 가지는 알고리즘은 존재할 수 없음이 증명되어 있다.
