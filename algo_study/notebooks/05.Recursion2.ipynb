조합의 수 계산

nCm 같은 combinations를 하는 것. 



```python
from math import factorial 

def combi(n, m):
    return factorial(n) / (factorial(m) * factorial(n-m))
```

```python
combi(9, 3)
```

원소 하나의 입장에서 

- 그 원소가 선택되었을 때 n-1개에서 m-1개만 뽑으면 되는 경우랑 
- 그 원소가 선택되지 않았을 때 n-1개에서 m개를 뽑으면 되는 경우

를 합친 것으로 보는 것이다. 





나의 방법

```python
def recursive_combi(n, m):
    if n < m:
        return 0
    
    if m == 1:
        return n
    
#     print(f'Combi({n-1}, {m-1}) + Combi({n-1}, {m})')
#     print(f'{recursive_combi(n-1, m-1)} + {recursive_combi(n-1, m)}')
    
    return recursive_combi(n-1, m-1) + recursive_combi(n-1, m)
```

suggested answer

```python
def recursive_combi(n, m):
    if n == m:
        return 1
    
    if m == 0:
        return 1   
    return recursive_combi(n-1, m-1) + recursive_combi(n-1, m)
```

```python
recursive_combi(9, 3)
```
